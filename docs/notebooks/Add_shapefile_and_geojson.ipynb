## new map instance

```python
from geohmt import Map,shp_to_geojson
```

```python
m = Map(center=[31.52,113.3],height="400px",zoom=12)
m
```

## shp transform geojson

```python
shp_path = r".\example\data\liulin_84.shp"
save_path = r".\example\data\liulin_84.geojson"
#".\example\data\liulin.shp"
shp_to_geojson(shp_path,savefile=save_path)
shp_to_geojson(shp_path)
```

## add shapefiles

```python
style = {
        "stroke" : True,
        "color" : "#000000",
        "weight" : 2,
        "opacity" : 1,
        "fill" : True,
        "fillColor" : "#000000",
        "fillOpacity" : 0.4,
    }
```

```python
m.add_shapefile(shp_path,style=style,layer_name="shpfile")
```

## add geojson

```python
m.add_geojson(save_path,style=style,layer_name="geojson")
```
